```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Start coding here! Use as many cells as you like
netflix_df[(netflix_df["release_year"] >= 1990) & (netflix_df["release_year"] < 2000) & (netflix_df["type"] == "Movie")]
plt.hist(netflix_df["duration"], bins=1000)
plt.show
duration = 90
print("The most frequent movie duration in the 1990s is approximately " + str(duration) + " mins.")
```

```python
# Action movies in the 90s with less than 90min duration
filtered_df = netflix_df[(netflix_df["release_year"] >= 1990) & (netflix_df["release_year"] < 2000) & (netflix_df["type"] == "Movie") & (netflix_df["genre"] == "Action")] # np.logical_and only accepts 3 conditions; thus &&

short_movie_count = 0

for index, row in filtered_df.iterrows(): # variable naming irrelevant; second variable accessible as row
    if row["duration"] < 90:
        short_movie_count += 1
        
print("There are " + str(short_movie_count) +" action movies released in the 1990s with a duration of less than 90mins in the dataset.")
```
